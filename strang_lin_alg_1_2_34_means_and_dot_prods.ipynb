```
import numpy as np
```

```
v = np.random.randn(3, 1)
u = v/np.linalg.norm(v)
u
```

```
V = np.random.randn(30, 3, 1)
U = np.zeros((30, 3, 1))
for i in range(len(V)):
  U[i] = V[i]/np.linalg.norm(V[i])

# Check if the for loop created unit vectors
for i in range(len(U)):
  print(np.linalg.norm(U[i]))
```

```
u_dot_U = np.dot(np.transpose(u), U)
u_dot_U
```

```
u_dot_U = u_dot_U.flatten()
u_dot_U
```

```
np.mean(u_dot_U)
```

```
np.var(u_dot_U)
```
